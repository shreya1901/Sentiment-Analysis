```python
import requests
from bs4 import BeautifulSoup

ratings_col = []
keywords_col = []
review_date_col = []
review_text_col = []

name = []
year = []
movie_rated = []
run_length = []
genres = []
release_date = []
IMDB_rating = []
num_raters = []
num_reviews = []
review_url =[]
image_Url = []


cnt = 0
for mv in movie_list:   
    cnt+=1
    print("{} : {}".format(cnt,mv[9]))
    page = requests.get(mv[9])
    soup = BeautifulSoup(page.content, "html.parser")
    
    job_elements = soup.find_all("div",{'class': 'lister-item-content'})
    i = 0; 
    for job_element in job_elements:
        i+=1
        name.append(mv[0])
        year.append(mv[1])
        movie_rated.append(mv[2])
        run_length.append(mv[3])
        genres.append(mv[4])
        release_date.append(mv[5])
        IMDB_rating.append(mv[6])
        num_raters.append(mv[7])
        num_reviews.append(mv[8])
        review_url.append(mv[9])
        image_Url.append(mv[10])
        rate = job_element.find("span",{'class': ''})
    #     print("{}".format(rate.text.strip()))
        if(len(rate.text.strip())<=2):
            ratings_col.append(int(rate.text))
    #         print(rate.text.strip())
        else:
            ratings_col.append(0)
    #         print(0)

        Keywords = job_element.find("a",{'class': 'title'})
#         i=i+1
        keywords_col.append(Keywords.text.strip())
#         print("{} : {}".format(i,Keywords.text.strip()))

        review_date = job_element.find("span",{'class': 'review-date'})
    #     i=i+1
        review_date_col.append(review_date.text.strip())
    #     print("{} : {}".format(i,review_date.text.strip()))

        review_text = job_element.find("div",{'class': 'text show-more__control'})
    #     i=i+1
        review_text_col.append(review_text.text.strip())
    #     print("{} : {}".format(i,review_text.text.strip()),end="\n"*2)

    #     print(job_element, end="\n"*2)
    
```
